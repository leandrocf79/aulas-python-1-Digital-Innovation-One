```python
# substituir valores nulos por zero
df["Vendas"].fillna(0, inplace=True)
```

```python
# apagar linhas com valores nulos
df.dropna(inplace=True)
```

```python
# apagar linhas com valores nulos com base apenas em 1 coluna, Vendas
df.dropna(subset=["Vendas"], inplace=True)
```

```python
# removendo linhas sem valores em todas as colunas
df.dropna(how="all", inplace=True)
```

```python
# criar coluna Receita de Vendas e Quantidade (mul significa multiplicar)
# exibir as 5 primeiras linhnas

df["Receita"]=df["Vendas"].mul(df["Qtde"])
df.head()
```

```python
# criar colunas Receita/Vendas e dividir uma pela outra. Isso volta ao valor anterior em float, apenas treinar. 
df["Receita/Vendas"] = df["Receita"] / df["Vendas"] 
df.head()
```

```python
# verificar a maior valor da receita
df['Receita'].max()
```

```python
# verificar a menor valor da receita
df['Receita'].min()
```

```python
# retornar os 5 maiores valores de Receita com nLargest
df.nlargest(5, 'Receita')
```

```python
# retornar os 7 menores valores de Receita com nsmallest
df.nsmallest(7, 'Receita')
```

```python
# agrupar por cidade em Receita
df.groupby("Cidade")["Receita"].sum()
```

```python
# converter coluna de Data em tipo inteiro
df['Data']=df['Data'].astype('int64')
df.dtypes
```

```python
# passar os dados da coluna Data para data
df['Data']=pd.to_datetime(df['Data'])
df.dtypes
```

```python
# agrupar por ano Data e Receita
df.groupby(df["Data"].dt.year)["Receita"].sum()
```

```python
# crianr nova coluna com o ano_venda atribuir Data
df['Ano_Venda']=df['Data'].dt.year
df.head()
```
